```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
import time
z= time.time()
x=print(func4([1,2,3,4] , [2,4,6,8]))
y= time.time()
print((y-z)*1000)
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
import time
z= time.time()
x=print(func5(16))
y= time.time()
print((y-z)*1000)
```

```python
# Exercise 6
def func8(A):
    return func8_helper(A, 0, len(A) - 1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - i + 1

    if n == 2:
        if A[i] > A[j]:
          # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
    elif n > 2:
        m = n // 3
        func8_helper(A, i, j - m)
        func8_helper(A, i + m, j)
        func8_helper(A, i, j - m)

    return A

```

```python
#Sample function call: func8([3,1,2,7,8,6])
import time
z= time.time()
x=print(func8([3,1,2,7,8,6]))
y= time.time()
print((y-z)*1000)
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
import time
z= time.time()
x=print(func12([1,2,3,4,5] , 2))
y= time.time()
print((y-z)*1000)
```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(A):
  total = 0
  for x in A :
    total = total + x
    return total
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def min_max(A):
  minimum = A[0]
  maximum = A[0]
  for x in A :
    if x < minimum :
      minimum = x
    if x > maximum:
      maximum = x
  return minimum , maximum
```
